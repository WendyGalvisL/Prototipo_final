```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# ============================================
# IMPORTAR K-MEDOIDS
# ============================================
from sklearn_extra.cluster import KMedoids
```

### 2. Preparación de Datos

```python
FEATURES = [
    "Servicios_Telecomunicaciones_No", "Servicios_Telecomunicaciones_Si",
    "Dispositivos_hogar_No", "conexion_hogar_si", "interrupciones_si",
    "frec_uso_si", "dens_int",
]

def load_data(path="data_procesada.csv"):
    df = pd.read_csv(path)
    for c in FEATURES:
        if c != "dens_int":
            df[c] = df[c].astype("float").astype("Int64").fillna(0).astype(int)
        else:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    if df["dens_int"].isna().any():
        df["dens_int"] = df["dens_int"].fillna(df["dens_int"].median())
    return df

def apply_weights(X: pd.DataFrame, weights: dict | None):
    Xw = X.copy()
    if weights:
        for col, w in weights.items():
            if col in Xw.columns:
                Xw[col] = Xw[col].astype(float) * float(w)
    return Xw

def scale_numeric(X: pd.DataFrame, numeric_cols=("dens_int",)):
    Xs = X.copy()
    scaler = StandardScaler()
    Xs[list(numeric_cols)] = scaler.fit_transform(Xs[list(numeric_cols)])
    return Xs

# Cargamos los datos
df = load_data("data_procesada.csv")
```

### 3. Función para K-Medoids

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score, adjusted_rand_score

def compute_internal_metrics(X: np.ndarray, labels: np.ndarray) -> dict:

    if len(np.unique(labels)) > 1:
        sil = silhouette_score(X, labels)
        ch = calinski_harabasz_score(X, labels)
        db = davies_bouldin_score(X, labels)
    else:
        sil, ch, db = np.nan, np.nan, np.nan
    return {"silhouette": sil, "calinski_harabasz": ch, "davies_bouldin": db}

def profile_sharpness(centroids: pd.DataFrame, binary_cols: list[str]) -> float:
    """Promedio de distancia del centroide a 0.5 (cuanto mayor, más 'nítido')."""
    if not binary_cols: return np.nan
    diffs = [np.abs(centroids[col] - 0.5) for col in binary_cols if col in centroids.columns]
    if not diffs: return np.nan
    return float(pd.concat(diffs, axis=1).mean().mean())

def cluster_size_ok(labels: np.ndarray, min_ratio=0.02) -> bool:
    n = len(labels)
    sizes = pd.Series(labels).value_counts(normalize=True)
    return (sizes >= min_ratio).all()

def centroid_table(X: pd.DataFrame, labels: np.ndarray):
    """Promedio por clúster (binarios → tasas; dens_int → media)."""
    dfc = X.copy()
    dfc["cluster"] = labels
    cent = dfc.groupby("cluster").mean(numeric_only=True).sort_index()
    sizes = dfc["cluster"].value_counts().sort_index().to_frame(name="n")
    sizes["pct"] = sizes["n"] / len(X)
    return cent, sizes


def stability_ari(X: np.ndarray, k: int, seeds: list[int], subsample=0.8) -> float:
    """Promedio de ARI entre múltiples corridas con submuestreo."""
    from sklearn_extra.cluster import KMedoids

    label_list = []
    rng = np.random.default_rng(42)
    n = X.shape[0]
    for seed in seeds:
        idx = rng.choice(n, size=int(n*subsample), replace=False)

        model = KMedoids(n_clusters=k, random_state=seed)
        labels = model.fit_predict(X[idx])
        label_list.append((idx, labels))

    if len(label_list) < 2: return np.nan
    aris = []
    for i in range(len(label_list)):
        for j in range(i+1, len(label_list)):
            idx_i, lab_i = label_list[i]
            idx_j, lab_j = label_list[j]
            common_idx = np.intersect1d(idx_i, idx_j, assume_unique=True)
            if len(common_idx) < 10: continue


            map_i = {val: pos for pos, val in enumerate(idx_i)}
            map_j = {val: pos for pos, val in enumerate(idx_j)}

            li = [lab_i[map_i[val]] for val in common_idx]
            lj = [lab_j[map_j[val]] for val in common_idx]

            aris.append(adjusted_rand_score(li, lj))
    return float(np.mean(aris)) if aris else np.nan

def geographic_entropy(df: pd.DataFrame, labels: np.ndarray, col_region="Nombre Departamento") -> float:
    """Entropía normalizada promedio de la distribución regional por clúster."""
    df_ = df[[col_region]].copy()
    df_["cluster"] = labels
    entropies = []
    for c, g in df_.groupby("cluster"):
        p = g[col_region].value_counts(normalize=True)
        H = -(p * np.log(p + 1e-12)).sum()
        H_max = np.log(len(p)) if len(p) > 1 else 1.0
        entropies.append(H / H_max)
    return float(np.mean(entropies)) if entropies else np.nan
```

```python
# FUNCIÓN PARA K-MEDOIDS
def run_kmedoids_grid(df: pd.DataFrame, k_values, weights=None, exp_name="exp"):
    # Preparación de datos
    X = df[FEATURES].copy()
    X = apply_weights(X, weights)
    X = scale_numeric(X, numeric_cols=("dens_int",))
    X_np = X.values

    results = []
    best = None

    for k in k_values:
        # Usa k-medoids
        km = KMedoids(n_clusters=k, init="k-medoids++", random_state=123)
        labels = km.fit_predict(X_np)

        m = compute_internal_metrics(X_np, labels)
        ok = cluster_size_ok(labels, min_ratio=0.02)

        # Los centroides en K-Medoids son los medoides (puntos reales)
        # La función centroid_table calcula promedios para perfilar el clúster
        cent, sizes = centroid_table(X, labels)
        ps = profile_sharpness(cent, [c for c in FEATURES if c != "dens_int"])

        # Se mide la estabilidad
        ari = stability_ari(X_np, k, seeds=[0,1,2,3,4], subsample=0.8)

        row = {
            "exp": exp_name, "k": k, "silhouette": m["silhouette"],
            "calinski_harabasz": m["calinski_harabasz"], "davies_bouldin": m["davies_bouldin"],
            "profile_sharpness": ps, "stability_ari": ari,
            "min_size_ok": ok, "sizes": sizes["pct"].round(3).to_dict()
        }
        results.append(row)

        # Lógica de selección del mejor clúster
        if ok:
            if best is None or (row["silhouette"] > best[0]["silhouette"]):
                best = (row, labels, cent, sizes)

    results_df = pd.DataFrame(results).sort_values("silhouette", ascending=False)
    return results_df, best
```

### 4. Ejecución de los Experimentos con K-Medoids

```python
# Experimento 1 — Línea base con K-Medoids

print("--- EJECUTANDO EXPERIMENTO 1 CON K-MEDOIDS ---")
k_values_exp1 = [4, 5, 6, 7, 8, 9, 10, 12]
res1_kmedoids, best1_kmedoids = run_kmedoids_grid(df, k_values_exp1, weights=None, exp_name="exp1_kmedoids")

print("Resumen EXP1 (K-Medoids):")
print(res1_kmedoids.head())
print("\nMejor resultado EXP1 (K-Medoids):", best1_kmedoids[0])


# Experimento 2 — Ponderación con K-Medoids

print("\n--- EJECUTANDO EXPERIMENTO 2 CON K-MEDOIDS ---")
weights_exp2 = {
    "Servicios_Telecomunicaciones_Si": 2.0,
    "conexion_hogar_si": 2.0,
    "frec_uso_si": 2.0, # (Asegúrate que esta columna exista o ajústala)
    "dens_int": 2.0
}
k_values_exp2 = [4, 5, 6, 7, 8]
res2_kmedoids, best2_kmedoids = run_kmedoids_grid(df, k_values_exp2, weights=weights_exp2, exp_name="exp2_kmedoids")

print("Resumen EXP2 (K-Medoids):")
print(res2_kmedoids.head())
print("\nMejor resultado EXP2 (K-Medoids):", best2_kmedoids[0])

# Experimento 3 — Estabilidad y Coherencia (con K-Medoids)

# Recuperar mejores resultados
# ('best1_kmedoids' y 'best2_kmedoids')
row1_kmedoids, labels1_kmedoids, cent1_kmedoids, sizes1_kmedoids = best1_kmedoids
row2_kmedoids, labels2_kmedoids, cent2_kmedoids, sizes2_kmedoids = best2_kmedoids


##  Lógica de selección del mejor modelo
def prefer(a, b):
    # Criterio 1: Silhouette
    if a["silhouette"] > b["silhouette"] + 1e-6: return "A"
    if b["silhouette"] > a["silhouette"] + 1e-6: return "B"

    # Criterio 2 Estabilidad ARI
    if a["stability_ari"] > b["stability_ari"] + 1e-6: return "A"
    if b["stability_ari"] > a["stability_ari"] + 1e-6: return "B"

    # Criterio 3 Nitidez de Perfil
    if a["profile_sharpness"] >= b["profile_sharpness"]: return "A"
    return "B"

## Comparar los mejores de cada experimento de K-Medoids
choice = prefer(row1_kmedoids, row2_kmedoids)

# Preparación de la configuración ganadora

if choice == "A":
    print("\n[EXP3] Partimos de la mejor configuración de EXP1 (K-Medoids)")
    # Recrear los datos usados en el Exp 1
    X_choice = scale_numeric(apply_weights(df[FEATURES], None), ("dens_int",))
    k_choice = row1_kmedoids["k"]
    labels_choice = labels1_kmedoids
    exp_winner_name = "EXP1"
else:
    print("\n[EXP3] Partimos de la mejor configuración de EXP2 (K-Medoids)")
    # Recrear los datos usados en el Exp 2 (con pesos)
    X_choice = scale_numeric(apply_weights(df[FEATURES], weights_exp2), ("dens_int",))
    k_choice = row2_kmedoids["k"]
    labels_choice = labels2_kmedoids
    exp_winner_name = "EXP2"

print(f"Configuración ganadora: {exp_winner_name} con k = {k_choice}")

## Validación final de estabilidad y coherencia
# Ejecutar una prueba de estabilidad más exhaustiva con 30 repeticiones

print("\nEjecutando validación final de estabilidad y coherencia...")
ari30 = stability_ari(X_choice.values, k_choice, seeds=list(range(30)), subsample=0.8)
geoH3 = geographic_entropy(df, labels_choice, col_region="Nombre Departamento")

print(f"Estabilidad (ARI promedio con 30 seeds): {ari30:.3f}")
print(f"Coherencia Territorial (Entropía Geográfica promedio): {geoH3:.3f}")

# Reporte final de perfiles y tamaños
# Perfiles finales (promedios por clúster) y los tamaños relativos

centF, sizesF = centroid_table(df[FEATURES], labels_choice)

print("\n--- PERFILES FINALES (MEDOIDES) ---")
print(centF.round(3))

print("\n--- TAMAÑOS FINALES DE CLÚSTERS ---")
print(sizesF.assign(pct = (sizesF["pct"]*100).round(1)))
```
